```python
import pandas as pd
from glob import glob
```

```python
files = glob("../data/2*.csv")
files = sorted(files)
files
```

```python
data = [pd.read_csv(f, skiprows=7, skipinitialspace=True) for f in files]
data = pd.concat(data)
data
```

```python
data = data[['GPS Reading', 'Lat (Degrees)', 'Lon (Degrees)', 'Velocity (MPH)', 'Heading(Deg)']].copy()
data['Velocity (KPH)'] = data['Velocity (MPH)'] * 1.60934
data.loc[data.index<400, 'Velocity (KPH)'] = 0 
data['Velocity (KPH)'] = data['Velocity (KPH)'].astype(int)
display(data['Velocity (MPH)'].describe())
display(data['Velocity (KPH)'].describe())
data
```

```python
data.to_csv("../data/combined.csv", index=False)
```
